```python
import bs4
```

```python
from urllib.request import urlopen as Ureq
```

```python
from bs4 import BeautifulSoup as soup
```

```python
my_url="https://www.flipkart.com/search?q=mouse&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

```python
uclient=Ureq(my_url)
```

```python
page_html=uclient.read()
```

```python
uclient.close()
```

```python
page_soup=soup(page_html,"html.parser")
```
